## Backup Models to Google Drive
Run this cell at the end to permanently save your work.

```python
from google.colab import drive
import shutil
import os

drive.mount('/content/drive')

drive_path = '/content/drive/MyDrive/NLP'
os.makedirs(drive_path, exist_ok=True) 

source_models = '/content/NLP-Assignmet/saved_models'
source_results = '/content/NLP-Assignmet/results'

if os.path.exists(source_models):
    shutil.copytree(source_models, f'{drive_path}/saved_models', dirs_exist_ok=True)
    print("\u2705 Models saved to Google Drive > NLP")
    
if os.path.exists(source_results):
    shutil.copytree(source_results, f'{drive_path}/results', dirs_exist_ok=True)
    print("\u2705 Results saved to Google Drive > NLP")
```
